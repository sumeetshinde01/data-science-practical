```python
import pandas as pd
import numpy as np
from scipy import stats
from statsmodels.stats import weightstats as stests
```

```python
df = pd.read_csv("/content/adult.csv.zip")
df.head()
```

1.One Sample t-test

```python
sample = df["age"].dropna()

t_stat, p_val = stats.ttest_1samp(sample, 40)

print("T-statistic:", t_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

Two Sample (Independent) t-test

```python
male = df[df["sex"]=="Male"]["hours.per.week"].dropna()
female = df[df["sex"]=="Female"]["hours.per.week"].dropna()

t_stat, p_val = stats.ttest_ind(male, female)

print("T-statistic:", t_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("Reject Null Hypothesis (Significant difference)")
else:
    print("Fail to Reject Null Hypothesis")
```

Paired t-test

```python
# create a paired variable (just for practice)
df["simulated_after"] = df["education.num"] + np.random.randint(-1, 2, len(df))

before = df["education.num"]
after = df["simulated_after"]

t_stat, p_val = stats.ttest_rel(before, after)

print("T-statistic:", t_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("Reject Null Hypothesis (significant difference)")
else:
    print("Fail to Reject Null Hypothesis")
```

Chi-Square test

```python
from scipy.stats import chi2_contingency

contingency = pd.crosstab(df["sex"], df["income"])

chi2, p, dof, expected = chi2_contingency(contingency)

print("Chi2 Statistic:", chi2)
print("P-value:", p)

if p < 0.05:
    print("Reject Null Hypothesis (Variables are related)")
else:
    print("Fail to Reject Null Hypothesis")
```

```python
groups = []

for level in df["education"].unique():
    group_data = df[df["education"] == level]["hours.per.week"].dropna()
    groups.append(group_data)

F_stat, p_val = stats.f_oneway(*groups)

print("F-statistic:", F_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("Reject Null Hypothesis (At least one group is different)")
else:
    print("Fail to Reject Null Hypothesis")
```

Z-test

```python
z_stat, p_val = stests.ztest(df["hours.per.week"].dropna(), value=40)

print("Z-stat:", z_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```
